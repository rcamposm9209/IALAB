**Ventajas**


1.   Son simples pero muy poderosos
2.   Caracterizan un problema mediante resultados de decisiones
3.   Capaces de manejar datos numéricos como categóricos
4. Requieren menos preparación los datos, con menor cantidad de suposiciones
5. Vinculados con la *regla de asociación* que a su vez son cercanas a modelos de razonamiento más complejo

**Desventajas**

1.   Tipicamente no alcanzan el mismo nivel de performance que otro métodos.
2.   No tiene buen desempeño cuando:

> Depende de muchos atributos
> Existen dependencias lineales/simple de representar de manera no discret

Aquí se usa solo como clasificación no como regresión.
